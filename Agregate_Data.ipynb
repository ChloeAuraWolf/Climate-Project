```python
import glob

file_paths = glob.glob("/home/central/majes1ce/notebooks/Climate Project/keycountsummary/*.csv")
```

```python
#Python class
class wordcounter:
    climate = 0
    precip = 0
    heat = 0
    events = 0
```

```python
from pathlib import Path
import csv

climate_keywords = {'climate', 'Climate', 'CLIMATE', 'change', 'CHANGE', 'Change', 'action', 'Action', 'ACTION'}
precip_keywords = {'rain', 'rains', 'Rain', 'Rains', 'RAIN', 'RAINS', 'rainfall', 'Rainfall', 'RAINFALL', 'stormwater', 'Stormwater', 'STORMWATER'}
heat_keywords = {'heat', 'Heat', 'HEAT', 'island', 'Island', 'ISLAND', 'stress', 'Stress', 'STRESS'}
events_keywords = {'weather', 'Weather', 'WEATHER', 'severe', 'Severe', 'SEVERE', 'Extreme', 'extreme', 'EXTREME', 'extremes', 'Extremes', 'EXTREMES'}

results = {}

for file_path in file_paths:
    filename = file_path.split("/")[-1]
    parts = filename.split('_',2)
    location = parts [0] + '_' + parts [1] #Shortens file name into county and city only
    
    results[location] = wordcounter()
    with open (file_path) as csvfile:
        df = csv.reader(csvfile)
        for row in df:
            if row[0] in climate_keywords:
                results[location].climate += int(row[1])
            if row[0] in precip_keywords:
                results[location].precip += int(row[1])
            if row[0] in events_keywords:
                results[location].events += int(row[1])
            if row[0] in heat_keywords:
                results[location].heat += int(row[1])

results = {key: value for key, value in sorted(results.items())}
                
print (f'Location,Climate Count,Precip Count,Events Count,Heat Count')  
for location, count in results.items():
    print(f'{location},{count.climate},{count.precip},{count.events},{count.heat}')
     
```
