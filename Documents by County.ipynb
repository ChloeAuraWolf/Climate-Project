```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

df = pd.read_excel("/home/central/majes1ce/notebooks/Climate Project/Book3.xlsx", sheet_name="Sheet1")


# Replace 'Variable1' and 'Variable2' with your actual column names
x = df['County']
y1 = df['Expected']
y2 = df['Actual']

# Plot the bar chart
x_indexes = np.arange(len(x))
width = 0.4  # Width of each bar

# Set up the figure and axis
plt.figure(figsize=(12, 6))

# Plot bars for each variable with different colors and shifted positions
plt.bar(x_indexes - width / 2, y1, width=width, color='skyblue', label='Expected', zorder=3)
plt.bar(x_indexes + width / 2, y2, width=width, color='salmon', label='Actual', zorder=3)
plt.xlabel('')
plt.ylabel("Number of Documents")
plt.title("Expected vs Actual Documents by County")
plt.xticks(ticks=x_indexes, labels=x, rotation=45, fontsize=6,ha='right')

plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.gca().yaxis.set_major_locator(ticker.MultipleLocator(5))

# Add legend to differentiate between variables
plt.legend()
plt.tight_layout()  # Adjust layout to prevent clipping
plt.savefig('E vs A')
plt.show()
```
